```python
import pandas as pd
import csv
```

```python
with open("data.csv","w") as csvfile: 
    writer = csv.writer(csvfile)
    writer.writerow(["dateType","timeQuantum","longitude","latitude","mesh","bikeNum","near_bikeNum","wind","temperature","weather"])
    with open("part-2") as f:
        for line in f.readlines():
            t = eval(line)
            writer.writerow(t[1])
```

```python
df = pd.read_csv("data.csv", header=0)
```

```python
with open("data_y.csv","w") as csvfile: 
    writer = csv.writer(csvfile)
    writer.writerow(["order"])
    with open("part-2") as f:
        for line in f.readlines():
            t = eval(line)
            writer.writerow([t[0]])
```

```python
y = pd.read_csv("data_y.csv", header=0)
```

```python
data = pd.concat([y, df], axis=1)
```

```python
data = data.drop(["timeQuantum"], axis=1)
```

```python
data.to_csv("oneday_part2.csv", index=False, float_format="%.16f")
```
