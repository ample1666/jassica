```python
import pandas as pd
import numpy as np
pd.set_option("precision", 16)  #显示的时候是16位的精度
```

```python
data = pd.read_csv("oneday_all.csv", header=0)
```

```python
data.head()
```

```python
data_6 = data[data.mesh == 6.0]
```

```python
data_6 = data_6.drop(["mesh"], axis=1)
```

```python
df_np = data_6.as_matrix()
df_np.shape
```

```python
y = np.log1p(df_np[:, 0])
X = df_np[:, 1:]
```

```python
X_train = X[:360000, :]
X_test = X[360000:, :]
y_train = y[:360000]
y_test = y[360000:]
```

```python
from xgboost import XGBRegressor
clf = XGBRegressor(max_depth=20)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
frank_y_test = np.exp(y_test) - 1
frank_y_pred = np.exp(y_pred) - 1
```

```python
d1 = pd.DataFrame(frank_y_test, columns=["test"])
d2 = pd.DataFrame(frank_y_pred, columns=["pred"])
d = pd.concat([d1, d2], axis=1)
d["test-pred"] = np.abs(d.test - d.pred)
d["ratio"] = d["test-pred"] / d.test
```

```python
d[d.ratio < 0.2]
```
